### ***Sorting and Subsetting***

```python
#Viene del más antiguo 
#Ordenar de mayor a menor, para visualizar la data. 
#Usando una columna númerica o stream(A,B,C,...)
#Se asigna a otra base de datos

net_old = netflix.sort_values("release_year")
net_old.head(10)
```

```python
net_new = netflix.sort_values("release_year", ascending = False)
net_new.head(10) #Muestra las primera 10 filas
#El ascending=False significa que no quiero ascending, es decir, quiero descendiente.
# En el caso anterio, es lo mismo poner ascending=True o dejarlo en blanco: siempre se asume True
```

```python
netflix['number'] = np.random.randint(1, 10, netflix.shape[0])
#Crear una columna de números aleatorios enteros entre 1 y 10: la cantidad de estos=cantidad de filas
```

```python
netflix.sort_values(["release_year","number"])
netflix.head(10)
#Ordenar por el ubigeo de provincias y por el numero
```

```python
net_two_sort = netflix.sort_values(["release_year","number"], ascending = [True,False]) 
net_two_sort.head(10)
#Ordenar de manera ascendente el release year y number de manera descendente
# obs : en 1943 ==> en ["duration"] lo ordena de fomra descendente
```

### ***Subsetting rows***

```python
net_50 = netflix["release_year"] < 2011
net_50
#Me da los indices de las peliculas que cumplen con la condición
```

```python
netflix[netflix["release_year"] < 2011].head(10)
#Me da los indices de las peliculas que cumplen con la condición
#Solo quiero que me imprima las 10 primeras filas, es a partir de los resultados 
#Sin .head(10) se tienen todos los resultados y no solo los 10 primeros
```

```python
netflix[(netflix["release_year"] < 2011) & (netflix["number"] > 5)]
#Me da los indices de las peliculas que cumplen con la condición 
#de ser antes de 2011 & que los numeros aleatorios sean mayores a 5
```

```python
netflix[( netflix.release_year < 2011 ) & (netflix.number > 5)]
#Es mejor que la condición de antes, menos engorroso en la escritura
```

# .isin()
If you want to filter by multiple values of a categorical variable, the easiest way is to use the isin() method.

```python
net_per_ch = netflix[netflix["country"].isin(["Peru","Chile"])]
net_per_ch
# filter by variables's values
#Filtrando por variables específicas
#Quiero que me vote "country" solo para caso de Peru y Chile.
```

```python
netflix[~netflix["country"].isin(["Peru","Chile"])]   
# ~ ALT + 126 
#Es lo opuesto a lo anterior
```

## Getting new data

```python
movie = netflix["type"] == "Movie"
m_usa = netflix["country"] == "United States"
m_actual = netflix["release_year"] > 2019

movie_usa = netflix[movie & m_usa & m_actual]
movie_usa
```

```python
movie_usa = netflix.loc[ ( netflix[ "type" ] == "Movie" ) & ( netflix[ "country" ] == "United States") & ( netflix[ "release_year" ] > 2019 ) ]
```

```python
movie_usa
```

```python
movie_usa.drop([ 'director'], axis=1 ) # drop columns has missing values

# axis = 0 (vertical), axis = 1 (horizontal)
#Quiero droppear en el horizontal, borrar columnas que tienen missing value
```

```python
movie_usa.dropna(axis=1)  # drop rows with missing values
#Si hay algun missing value me va a borrar la fila, ya que es (axis=1)
#Coge el nombre de variables
```

```python
movie_usa.dropna(axis=0) # drop columns with missing values
#Coge a los indexis (es una operacion vertical)  
```

```python
movie_usa[movie_usa.release_year != 2020]
#No quiero que aparezca el 2020
```

```python
movie_usa.rename(columns = {'title':'Titulo_movie', 'duration':'Duration_movie'}, inplace = True)

#PARA CAMBIAR NOMBRE DE VARIABLES EN PYTHON
# rename (var1 var2) (var3 var4) en STATA
```
