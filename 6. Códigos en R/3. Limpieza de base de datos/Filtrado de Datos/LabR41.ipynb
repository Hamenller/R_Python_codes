### ¿A qué sexo pertenecen el mayor número de fallecidos?

```r
#Cantidad de fallecimientos segun genero
#nrow(data[se añade el condicional])


nrow(covid[covid$SEXO == "MASCULINO",])


nrow(covid[covid$SEXO == "FEMENINO",])
```

```r
#Es lo mismo si pongo
nrow(covid[SEXO == "MASCULINO",])


nrow(covid[SEXO == "FEMENINO",])
```

```r
#Se hace lo mismo pero usando group_by (lo del collap): voy a colapsar, quiero que me agrupe por femenino y masculino y queiro que me cuentes
# count using dplyer library : contar por genero
#Hasta ahora solo se esta agrupando
covid %>% group_by(SEXO)
```

```r
#Aqui ya se cuenta cuanto se tiene de masculino y fememnino
#count rows
#vean que la variable sexo hay observaciones en donde no solo estan los valores femenino y masculino, sino dos valores numericos
#Lo cierto es que councide con la funcion anterior
covid %>% group_by(SEXO) %>% tally()
```

```r
# Replacing values [using paquete dplyr]: para ello se usa mutate (tambien se usa para crear variables, para cambiar nombres de variables, para reemplazar valroes)
#Haciendo reemplazos y modificaciones
#Se reemplazan valores: 10 y 11 coloca NA
#Mutate: voy a reemplazar valores numericos en la variable sexo; es decir, va a ser igual a la variable sexo igual, pero los valores 10 y 11 se van a reemplazar por NA.
#Luego a este cambio, a esta base de datos alterada, lo voy a agrupar y luego a esta base agrupada lo voy a contar cuanto femenino y masculino tengo.

covid %>% mutate(SEXO = replace(SEXO, (SEXO == 10 | SEXO == 11) , NA))  %>% group_by(SEXO) %>% tally()
```

```r
#Haciendo varias modificaciones a la base en una sola linea de codigo: no se puede en Python
#filter + replace values + collapse(group_by)

covid %>%  filter(EDAD_DECLARADA > 30 & EDAD_DECLARADA < 60) %>% 
           mutate(SEXO = replace(SEXO, (SEXO == 10 | SEXO == 11) , NA))  %>% 
            group_by(SEXO) %>% tally()

```


## <a id='def'>Halla el total de fallecidos por departamento.</a>

```r
#Generar una base de datos donde saque la cantidad de personas fallecidas por departamento (collapse), las voy a contar, pero tambien voy a renombrar la variable
#Renombro: ya no será "n" como es habitualemnte, sino "deaths_departamento"
#A este resultado le voy a decir que se llame "departamento"

departamento <- covid %>% group_by(DEPARTAMENTO) %>% tally %>% rename(deaths_departamento = n)
```

```r
departamento
```

```r
### Halla el total de fallecidos por provincia, se debe agrupar por departamento y provincia, porque hay provincias que tienen el mismo nombre, que pertenecen a diferentes departamentos. 
#Peor aun con distritos. Luego a ese dato le voy a cambiar el nombre como "deaths_provincia"

provincia <- covid %>% group_by(DEPARTAMENTO, PROVINCIA) %>% 
                tally %>% 
                rename(deaths_provincia = n)
```

```r
provincia
```

```r
head(provincia)
```

```r
#Contabilizar el número de personas fallecidas por distrito.
#Se debe tomar en cuenta que hay distritos que tienen el mismo nombre en varios departamentos.
#Para encontrar ese identificador único de distrito, se tiene que agrupar por: DEPARTAMENTO, PROVINCIA, DISTRITO
#no se tiene el problema que se tiene en Python (indexing)
distritos <- covid %>% group_by(DEPARTAMENTO, PROVINCIA, DISTRITO) %>% tally %>% rename(deaths_distrito = n)
```

```r
head(distritos)
```

## <a id='def'>Merge Dataset </a>

```r
#En el Merge hay una base que es la "master data_set": mi base principal
#A esa base principal le agrego la informacion de otra base 
#La otra base de datos en la "using data_set"
#STATA m:1
```

```r
#Se va a mergear, se tiene que basar en las variables interesadas
#mi master data es distritos y a este le quiero añadir la informacion de provincia, pero en esta base de datos aparece la provincia una vez 
#En cambio, en mi base princiapal, distritos, aparece la provincia más de una vez, porque dentro de provincias tengo distritos.
#Entonces, para hacer el merge, dado que quiero que me aparezca la informacion por provincia. Por ello, se va a tomar en cuenta a ("DEPARTAMENTO", "PROVINCIA")
#Cuando se le coloca all.x = TRUE, ello significa que voy a elegir a la master data a (x,y)=(distritos, provincia). Entonces, distritos será mi base brinciapal o master data.

base_total <- merge( distritos, provincia , by = c("DEPARTAMENTO", "PROVINCIA"), all.x = TRUE)
```

```r
base_total

#Se repite la informacion de las provincias para cada distrito
```

```r
#Al agregar departamento, se repite la informacion de los departamentos para cada distrito.
#Merge de departamentos con la nueva base principal (ya no es distritos, sino es base_Total dado que ses la modificacion de la primera)
#Uno las bases, la using data es departamento

base_total <- merge( base_total, departamento, by = c("DEPARTAMENTO"), all.x = TRUE)
```

```r
head(base_total)
```

```r
#Lo que al final quiere es sacar un porcentaje:
#Porcentaje de personas fallecidas por covid en cada distrito en proporcion a la cantidad de personas fallecida en la provincia en la cual pertenece el distrito
#Porcentaje de personas fallecidas por covid en provincias en proporcion a la cantidad de personas fallecida en el departamento en la cual pertenece el distrito
#Porcentaje de personas fallecidas por covid en cada distrito en proporcion a la cantidad de personas fallecida en el departamento en la cual pertenece el distrito

#Para ponerle dos decimales se usa (digit=2)
```

```r
base_total$distr_prov = round(((base_total$deaths_distrito)/(base_total$deaths_provincia))*100, digit=2)
base_total$prov_dpto = round(((base_total$deaths_provincia)/(base_total$deaths_departamento))*100, digit=2)
base_total$distr_dpto = round(((base_total$deaths_distrito)/(base_total$deaths_departamento))*100, digit=2)


#muertos_covid['distr_prov']=(muertos_covid['Muertos por distrito']/muertos_covid['Muertos por provincia'])*100
#muertos_covid['prov_dpto']=(muertos_covid['Muertos por provincia']/muertos_covid['Muertos por departamento'])*100
#muertos_covid['distr_dpto']=(muertos_covid['Muertos por distrito']/muertos_covid['Muertos por departamento'])*100
#Tabla_1 = muertos_covid[['DEPARTAMENTO','PROVINCIA','DISTRITO','distr_prov']].round(2)
#Tabla_1
```
